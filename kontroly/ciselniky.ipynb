```python
import pandas as pd
import os
```

```python
if not os.path.exists("cache"):
    os.makedirs("cache")
if not os.path.exists("cache/ciselniky.csv"):
    ciselniky = pd.read_csv("https://ws.ujep.cz/ws/services/rest2/ciselniky/getCiselnikNewDomains?outputFormat=CSV&outputFormatEncoding=utf-8")
    ciselniky.to_csv("cache/ciselniky.csv", index=False)
```

```python
ciselniky = pd.read_csv("cache/ciselniky.csv")
ciselniky_data = {}
REFRESH_CISELNIKY = True
for ciselnik in ciselniky["item"].unique():
    if os.path.exists("cache/ciselniky/" + ciselnik + ".csv") and not REFRESH_CISELNIKY:
        continue
    print("fetching data for", ciselnik)
    try:
        data = pd.read_csv("https://ws.ujep.cz/ws/services/rest2/ciselniky/getCiselnik?domena={}&outputFormat=CSV&outputFormatEncoding=utf-8".format(ciselnik), sep=";", dtype=str)
        if data.shape[0] == 0:
            raise Exception("empty data")
        ciselniky_data[ciselnik] = data
    except Exception as e:
        print("couldn't fetch data for", ciselnik)
```

```python
if not os.path.exists("cache/ciselniky"):
    os.mkdir("cache/ciselniky")
for ciselnik, data in ciselniky_data.items():
    path = "cache/ciselniky/{}.csv".format(ciselnik)
    if os.path.exists(path):
        os.remove(path)
    data.to_csv(path, index=False, encoding="utf-8", sep=";")
```
